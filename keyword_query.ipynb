```python
import os
import pandas as pd
```

```python
# Laita kaikki datafilet yhteen kansioon ja muuta polku vastaamaan sitä kansiota
data_files = os.listdir("data/")
```

```python
# Haetaan data-kansiossa olevien tiedostojen nimet
files = [f for f in data_files if not f.startswith('.')]
files = sorted(files)
```

```python
# Otetaan ensin talteen datassa olevien sarakkeiden nimet
with open("data/" + files[0], "r") as f_lines: # Avataan ensimmäinen tiedosto
    columns = f_lines.readline() # Luetaan sarakkeiden nimet, nämä samat kaikissa tiedostoissa

print(columns) # Tarkistetaan
```

```python
# Lista johon kerätään paprujen tiedot
data = []

# Avataan kukin tiedosto, käydään läpi riveittäin ja lisätään datat listalle
for f in files:
    
    with open("data/" + f, "r") as f_lines: # Avataan tiedosto
        
        f_lines.readline() # Heitetään pois sarakkeiden nimet eli tiedoston ensimmäinen rivi
        
        for l in f_lines: # Käydään läpi tiedoston loput rivit
            
            # Rikotaan rivi listaksi ja lisätään data-listalle
            # split()-kutsun sisällä on WoS:n omituinen merkki joka erottelee tietueet
            data.append(l.split('	'))
```

```python
# Muutetaan pandas-dataframeksi, jonne helppo tehdä hakuja
df = pd.DataFrame(data, columns=columns.split("	"))
df.columns = [c.strip().replace("\ufeff", "") for c in df.columns] # Siivotaan columnien nimet
```

```python
# Kattellaa vähä
df.head()
```

```python
# Kirjoitetaan koko data csvksi
df.to_csv('data_full.csv', index=False)
```

```python
# Isossa osassa dataa ei näytä olevan abstrakteja
print("Papruja joissa ei absuja:", len(df[df['AB'] == '']), "/", len(df))
```

```python
# Papruja joiden titlessä on string "behavio"
df['TI'].str.contains("behavio", case=False).value_counts()
```

```python
# Papruja joiden abstraktissa on string "behavio"
df['AB'].str.contains("behavio", case=False).value_counts()
```

```python
# Valitaan nää paprut
mask_title = df['TI'].str.contains("behavio", case=False)
mask_abs = df['AB'].str.contains("behavio", case=False)

df_behavio = df[mask_title | mask_abs]
```

```python
df_behavio.reset_index(inplace=True, drop=True)
df_behavio
```

```python
# Kirjotetaan csv
df_behavio.to_csv("data_behavio.csv", index=False)
```
