### Abstracts

```python
df = pd.read_csv("data_scopus/aglr.csv")
```

```python
df['abstract'] = ''
eids = df['eid']
```

```python
for i in range(0, len(df)):
    ab = AbstractRetrieval(eids[i], view='FULL', id_type='eid')
    
    print(i, end="\r")
    
    if ab.abstract != None:
        df.loc[i, 'abstract'] = ab.abstract
    else:
        df.loc[i, 'abstract'] = ab.description
```
